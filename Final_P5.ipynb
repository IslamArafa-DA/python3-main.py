```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

def draw_plot():
    # Read data from file
    df= pd.read_csv('epa-sea-level.csv')

    # Create scatter plot
    plt.figure(figsize=(18,10))
    plt.scatter(x= df['Year'],y=df['CSIRO Adjusted Sea Level'],color='blue',label=' Sea Level')

    #### Create first line of best fit
    
    # Extract relevant columns
    x = df['Year']
    y = df['CSIRO Adjusted Sea Level']

    # Perform linear regression
    slope, intercept,_,_,_ = linregress(x, y)

    # Print the results
    print(f"Slope: {slope}")
    print(f"Intercept: {intercept}")

    # Create the line of best fit
    line_x = range(int(x.min()), 2050 + 1)  # Generate x values for the line Extended to 2050
    line_y = [slope * xi + intercept for xi in line_x]  # Calculate corresponding y values
    
    # Predict sea level for 2050
    sea_level_2050 = slope * 2050 + intercept
    print(f"Predicted Sea Level in 2050: {sea_level_2050} inches")
    
    # Plot the line of best fit
    plt.plot(line_x, line_y, color='red', label='Sea level to 2050')

    #### Create second line of best fit
    
    # Filter data from the year 2000 onward
    N_df = df[df['Year'] >= 2000]
    
    # Extract relevent columns
    xn = N_df['Year']
    yn = N_df['CSIRO Adjusted Sea Level']
    
    # Perform linear regression
    slope_N, intercept_N,_,_,_ = linregress(xn, yn)
    
    # Print the results
    print(f"Slope_N2000: {slope_N}")
    print(f"Intercept_N2000: {intercept_N}")
    
    # Create the New line of best fit
    line_xn = range(2000, 2050 + 1)  # Generate x values for the New line from 2000 Extended to 2050
    line_yn = [slope_N * xni + intercept_N for xni in line_xn]  # Calculate corresponding y values
    
    # Predict sea level for 2050 for The New line
    N_sea_level_2050 = slope_N * 2050 + intercept_N
    print(f"Predicted Sea Level in 2050: {N_sea_level_2050} inches")
    
    # Plot the New line of best fit
    plt.plot(line_xn, line_yn, color='green', linestyle = '--',label='N_Sea level to 2050')

 
    # Add labels and title
    plt.title('Rise in Sea Level',fontsize = 20)
    plt.xlabel('Year', fontsize= 18)
    plt.ylabel('Sea Level (inches)',fontsize= 18)
    plt.legend()
    plt.grid(True)

    
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```

```python
draw_plot()
```
